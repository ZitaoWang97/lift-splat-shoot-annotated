```python
import torch
```

```python
# grid_confc 网格的配置
```

```python
xbound=[-50.0, 50.0, 0.5]  # 限制x方向的范围并划分网格
ybound=[-50.0, 50.0, 0.5]  # 限制y方向的范围并划分网格
zbound=[-10.0, 10.0, 20.0]  # 限制z方向的范围并划分网格
dbound=[4.0, 45.0, 1.0]  # 限制深度方向的范围并划分网格
```

```python
def gen_dx_bx(xbound, ybound, zbound):  # 划分网格
    dx = torch.Tensor([row[2] for row in [xbound, ybound, zbound]])  
    bx = torch.Tensor([row[0] + row[2]/2.0 for row in [xbound, ybound, zbound]])  
    nx = torch.LongTensor([(row[1] - row[0]) / row[2] for row in [xbound, ybound, zbound]])  
    return dx, bx, nx
```

```python
dx, bx, nx = gen_dx_bx(xbound, ybound, zbound)
```

```python
# 分别为x, y, z三个方向上的网格间距
dx 
```

```python
# 分别为x, y, z三个方向上第一个格子的坐标
bx
```

```python
# 分别为x, y, z三个方向上格子的数量
nx
```

```python
# create_frustum 创建视锥
```

```python
# 原始图片大小
ogfW = 352
ogfH = 128
# 下采样倍数为16
fH = 8
fW = 22
# D: 41 表示深度方向上网格的数量
D = 41  
# 在0到351上划分22个格子 xs: DxfHxfW(41x8x22)
xs = torch.linspace(0, ogfW - 1, fW, dtype=torch.float).view(1, 1, fW).expand(D, fH, fW)
# 在0到127上划分8个格子 ys: DxfHxfW(41x8x22)
ys = torch.linspace(0, ogfH - 1, fH, dtype=torch.float).view(1, fH, 1).expand(D, fH, fW)
```

```python
ds = torch.arange(*dbound, dtype=torch.float).view(-1, 1, 1).expand(-1, fH, fW)
```

```python
frustum = torch.stack((xs, ys, ds), -1)
```
